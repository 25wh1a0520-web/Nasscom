```python
fruits = ['apple','banana','cherry','apple']
print("orginal list:",fruits)
print("list fruits:",fruits[0])
```

```python
fruits = ['apple','banana','cherry','apple']
fruits.append('orange')
print("after adding append:",fruits)
```

```python
fruits = ['apple','banana','cherry','apple']
fruits.remove('banana')
print("after removed banana:",fruits)
print("number of fruits:",len(fruits))
```

```python
unique_numbers = {1,2,3,2,1}
print("set(noduplicates):",unique_numbers)
```

```python

set_a = {1, 2, 3}
set_b = {3, 4, 5}
print("Union:", set_a | set_b)
print("Intersection:", set_a & set_b)
print("Difference (A - B):", set_a - set_b)

```

```python
student = {
    'name': 'Alice',
    'age': 21,
    'course': 'Python'
}
print("Dictionary:", student)
```

```python
print("Student name:", student['name'])
```

```python
student['grade'] = 'A'
print("After adding grade:", student)


del student['age']
print("After deleting age:", student)


for key, value in student.items():
    print(f"{key}: {value}")
```

```python
keys = ['id', 'name', 'marks']
values = [101, 'John', 88]
student_info = dict(zip(keys, values))
print("Student Info from zip:", student_info)

```
